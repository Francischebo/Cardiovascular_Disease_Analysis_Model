```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory


# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session

df= pd.read_csv('heart.csv')
print(df)


df.head(5)

df.columns

df['target'].hist(color='C4')

df.hist(bins=50,figsize=(20,15),color='C4')

corr_matrix=df.corr()
corr_matrix['target'].sort_values(ascending=False)


from pandas.plotting import scatter_matrix
attributes=['target','age', 'sex', 'cp',
       'exang', 'oldpeak', 'slope', 'ca', 'thal']
scatter_matrix(df[attributes],figsize=(12,8),color="C6")


df.describe()

df.info()

df.shape

X=df.drop('target',axis=1)

X

y=df['target']

```

```python
from sklearn.model_selection  import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2)



from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_train_scaled=scaler.fit_transform(X_train.astype(np.float32))





from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier



sgd_clf=SGDClassifier(random_state=42)
rnd_clf=RandomForestClassifier()
svc_clf=SVC()
log_reg = LogisticRegression()
tree_clf=DecisionTreeClassifier()

models=[sgd_clf,rnd_clf,svc_clf,log_reg,tree_clf]
for model in models:
    accuracy= cross_val_score(model,X_train_scaled,y_train,cv=3,scoring='accuracy')
    print(f"model name : {model} and it's accuracy:{accuracy}")
    
    
log_reg.fit(X_train_scaled,y_train)


y_pred_train= log_reg.predict(X_train_scaled)


from sklearn.metrics import confusion_matrix
confusion_matrix(y_train, y_pred_train)


from sklearn.metrics import precision_score, recall_score

recall_score(y_train,y_pred_train),precision_score(y_train,y_pred_train)




from sklearn.metrics import f1_score

f1_score(y_train,y_pred_train)



from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_test_scaled=scaler.fit_transform(X_test.astype(np.float32))




y_pred=log_reg.predict(X_test_scaled)



confusion_matrix(y_test, y_pred)



f1_score(y_test,y_pred)
```
